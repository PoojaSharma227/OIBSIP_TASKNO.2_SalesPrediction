*Segregate Dataset into X & Y*

**SPLITTING DATASET INTO TRAIN & TEST**

```python
from sklearn.model_selection import train_test_split
X = df[['TV', 'Radio', 'Newspaper']]
Y = df['Sales']
X_train,X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.30, random_state = 0)
print(X_train.shape)
print(X_test.shape)
```

**TRAINING**

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print(model.intercept_)
print(model.coef_)
```

**PREDICTION**

```python
y_pred = model.predict(X_test)
```

```python
act_predict = pd.DataFrame({'Actual': y_test.values.flatten(), 'Predict': y_pred.flatten()})
act_predict.head(10)
```

```python
sns.lmplot(data= act_predict, x='Actual', y='Predict')
```

**EVALUTION OF THE MODEL**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print("Mean_Absolute_Error:", mean_absolute_error(y_test, y_pred))
print("Mean_Squared_Error:", mean_absolute_error(y_test, y_pred))
print("Square_Mean_Absolute_Error:", np.sqrt(mean_absolute_error(y_test, y_pred)))
print("r2_score:", r2_score(y_test, y_pred))
```

since this problem is a regression problem as our dataset is not continous. so hence we select linear regression as a model for this sale prediction.
